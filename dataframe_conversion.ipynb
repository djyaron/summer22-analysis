Next, we define a simple helper function to calculate whether a given residual for a molecular configuration is an outlier based on quantiles.

```python
from typing import Union

from pandas import DataFrame, Series


def is_outlier(
    x: Union[DataFrame, Series], q_lower: float = 0.25, q_upper: float = 0.75
) -> Union[DataFrame, Series]:
    return (x < x.quantile(q_lower)) | (x > x.quantile(q_upper))
```

But first, to understand what this is doing, let's see what `.quantile()` returns on a dataframe:

```python
resid.quantile(0.25)
```

So, the `.quantile()` method computes the quantile for each column in the dataframe, returning a Series.

Now, we can use this output to compare against the original dataframe, to identify all the molecular configurations which have a residual less than the 25% quantile:

```python
resid < resid.quantile(0.25)
```

The reason we're allowed to use the less than operator (`<`) between `resid`--a DataFrame with many columns--and `resid.quantile(0.25)`--a Series, with only one column, is because of something called "broadcasting". You can think of what's happening as Pandas transposing the `resid.quantile(0.25)` Series into a row with 91 columns, and then performing an element-wise comparison with each row in the `resid` dataframe (think matrix-vector multiplication).

Now that we have all the pieces, let's write out exactly what the `is_outlier` function is doing so we understand it fully:

```python
(resid < resid.quantile(0.25)) | (resid > resid.quantile(0.75))
```

Let's use the is_outlier function and Pandas operations to compute some statistics on the outliers.

Below, we compute the frequency of outlier appearance for each molecular configuration:

```python
is_outlier(resid).sum(axis="columns")
```

Let's also calculate the percentage of a molecule's configurations that are marked outliers, for each method-method combination.

```python
is_outlier(resid).groupby(level=[0]).mean()
```

This works because we tell Pandas to group only by the first level of the MultiIndex--the "name" level. This tells Pandas to calculate the mean over all the individual configurations of each molecule.

As an aside, we make use of the fact that True is interpreted as 1 and False as 0 when we use numeric methods on boolean values. For example the mean of [True, True, True, False] is 0.75.
